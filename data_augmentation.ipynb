```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img
import os
import numpy as np
import imageio

datagen = ImageDataGenerator(
    
    rotation_range=90,
    brightness_range=[0.2,1.0],
    shear_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest')


# Assuming your images are loaded in train_ds variable
for batch, labels in train_ds:
    images = batch.numpy()
    labels = labels.numpy()

    for idx, image in enumerate(images):
        label = labels[idx]
        augmented_images = datagen.flow(image.reshape((1,)+image.shape), batch_size=32)
        if label == 0:
            label = "defective"
        else:
            continue
        i = 0
        for aug_image in augmented_images:
            aug_image = array_to_img(aug_image[0])
            # Save each augmented image with a name that includes the label
            aug_image.save(f"dataset/bottle/test/augmented/aug_{label}_{i}_{np.random.randint(0, 100000)}.png")
            i += 1
            if i > 4:  # otherwise the generator would loop indefinitely
                break
```
